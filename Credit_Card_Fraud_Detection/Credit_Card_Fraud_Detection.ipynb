```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
credit_data = pd.read_csv('creditcard.csv')
```

```python
# Now data counts on different classes
credit_data.value_counts('Class')
```

```python
# sperateing the dataset based on the class
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
# Analysing data based on the amount withdraw
legit.Amount.describe()
```

```python
fraud['Amount'].describe()
```

```python
# grouping the based on the class
credit_data.groupby('Class').mean()
```

```python
# for the blancing the dataset,
new_legit = legit.sample(n = 500 , axis = 0)
```

```python
new_legit.value_counts('Class')
```

```python
new_legit.describe()
```

```python
# let concanate legit and fraudent dataset
data = pd.concat([new_legit,fraud])
```

```python
data.describe()
```

```python
data.shape
```

```python
data['Class'].value_counts()
```

```python
data.groupby('Class').mean()
```

```python
# now let sort the data on X and Y
X = data.drop('Class',axis = 1)
Y = data['Class']
```

```python
print(X.shape,Y.shape)
```

```python
# SPlit the data into train and test perpotion of the class
X_train,X_test,Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify  =Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Now we have our train and test
# loading the the model for training
classifier = LogisticRegression()
```

```python
# fit the train data
classifier.fit(X_train,Y_train)
```

```python
# checking accurary on train data
X_train_prediction = classifier.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction,Y_train)
print(X_train_accuracy)
```

```python
# confusion matix on train data
conf_train = confusion_matrix(X_train_prediction,Y_train)
print(conf_train)
```

```python
# checking accurary on train data
X_test_prediction = classifier.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction,Y_test)
print(X_test_accuracy)
```

```python
# confusion matix on test data
conf_test = confusion_matrix(X_test_prediction,Y_test)
print(conf_test)
```

```python
# now let's make prediction for the new data set
input_data = credit_data.iloc[280149].drop('Class')
input_data_arry = np.asarray(input_data)
reshaped_data = input_data_arry.reshape(1,-1)
predition = classifier.predict(reshaped_data)
print(predition)
```
